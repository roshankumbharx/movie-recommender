```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies.loc[:, 'genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies.loc[:, 'keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter<3:
            L.append(i['name'])
            counter+=1
        else:break
    return L

movies.loc[:,'cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
movies.loc[:,'crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview']
```

```python
movies=movies.dropna()
```

```python
movies.isnull().sum()
```

```python
movies.loc[:,'overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies.loc[:,'genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies.loc[:,'keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies.loc[:,'cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies.loc[:,'crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['genres']+movies['overview']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df=movies[['movie_id','title','tags']]
new_df
```

```python
new_df[new_df['movie_id'].duplicated()]
```

```python
new_df.drop_duplicates(subset='movie_id',keep='first',inplace=True)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
# performing stemming
!pip install nltk
```

```python
import nltk

from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
pip install -U scikit-learn
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
vectors
```

```python
vectors[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity =  cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
import sys
```

```python
print(sys.getsizeof(similarity))
```

```python
upper_matrix_indices = np.triu_indices_from(similarity)

upper_matrix_indices
```

```python
upper_matrix_values = similarity[upper_matrix_indices]
print(sys.getsizeof(upper_matrix_values))
```

```python
pickle.dump(upper_matrix_indices,open('upper_mat_ind.pkl','wb'))
```

```python
reconstructed_mat=np.zeros((4800,4800))
reconstructed_mat[upper_matrix_indices]=upper_matrix_values
reconstructed_mat = reconstructed_mat + reconstructed_mat.T - np.diag(reconstructed_mat.diagonal())
```

```python
def recommend(movie):
    movie_index =  new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list= sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)
        
```

```python
recommend('Dead Poets Society')
```
